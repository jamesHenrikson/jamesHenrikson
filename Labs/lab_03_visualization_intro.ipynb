```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

## Preview Data

Use the following data dictionary to understand the attributes available

|Attribute | Description |
|---------|--------------|
|	mpg	|Miles/(US) gallon|
|	cyl|	Number of cylinders|
|	disp|	Displacement (cu.in.)|
|hp|	Gross horsepower|
|drat|	Rear axle ratio|
|wt|	Weight (1000 lbs)|
|qsec|	1/4 mile time|
|vs|	Engine (0 = V-shaped, 1 = straight)|
|am|	Transmission (0 = automatic, 1 = manual)|
|gear|	Number of forward gears|


# Create basic charts

- Line chart
- Bar chart
- Histogram
- Box plot
- Scatter plot




```python
#Line chart
df.plot(x='carmodel', y='mpg', kind='line', title="MPG by Car")
```

```python
#Bar chart
df.plot(x='carmodel', y='mpg', kind='bar', title="MPG by Car")
```

```python
#Histogram
df['mpg'].plot(kind='hist', bins=10, title="MPG Distribution")
```

```python
#Boxplot
df['mpg'].plot(kind='box', title = 'Miles Per Gallon')
```

```python
#Scatter plot
df.plot.scatter(x='mpg',y='hp',c='wt')
```

🔧 Assignment Tasks
- 1.1 Create a pie chart showing the count of cars by transmission type.  Display labels showing the percentage of each transmission type.  
- 1.2 Create a bar chart of cars and HP, with different colors for each car model
- 1.3 Create a scatter plot of wt (weight) vs. mpg.
- 1.4 Create a bar chart showing the average mpg for each cyl.
- 1.5 Plot a histogram of the weight column.

```python
#Pie chart 🔧

# Combine value_counts, rename for labels, and plot.pie into a single line
ax = df['am'].value_counts().rename(index={0: 'Automatic', 1: 'Manual'}).plot.pie(autopct='%1.1f%%', figsize=(7, 7), title='Distribution of Transmission Types')
ax.set_ylabel('') # Hide the default y-label
```

```python
#Bar chart 🔧
# Generate a bar chart of cars and HP, with different colors for each car model
# Using a colormap that provides more distinct colors for a larger number of items
import numpy as np
cmap = plt.colormaps['gist_rainbow']
colors = [cmap(i) for i in np.linspace(0, 1, len(df))]
df.plot(x='carmodel', y='hp', kind='bar', title='HP by Car Model', color=colors, legend=False)
```

```python
#Scatter plot 🔧
df.plot.scatter(x='wt',y='mpg')
```

```python
#Bar chart 🔧
bar_chart = df.groupby('cyl')['mpg'].mean().plot(kind='bar', title='Average MPG by Cylinder')
bar_chart.set_xlabel('Cylinders')
bar_chart.set_ylabel('Average MPG')
```

```python
#Histogram 🔧
df['wt'].plot(kind='hist', bins=10, title="Weight Distribution")
```
